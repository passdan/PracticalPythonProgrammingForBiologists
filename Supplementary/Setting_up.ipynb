## Course Preparation
You can follow these course materials with any python installation, but for this course we're going to use VScode and the google colab environment to ensure we're all using the same tools. Later in the course we will run python on our own computers, but as there can be issues with instalations, admin permissions, windows vs macs etc., it's good to start first with something we know works for everyone!

### Fork the Repository
Firstly lets get the code we're going to be working on for this course

- Open VS Code
- Press Ctrl+Shift+P (or Cmd+Shift+P) and type "Git: Clone"
- Enter the repository URL: https://github.com/passdan/PracticalPythonProgrammingForBiologists.git
- Choose a local folder to clone into
    - When prompted "Would you like to open the cloned repository?", click "Open"

Now lets create your own copy of the code that you can edit and save:
- Press Ctrl+Shift+P and type "GitHub: Fork"
- Select "Fork the current repository"
- Choose your GitHub account as the destination

### Set Up Google Colab in VS Code
Your computer may already have python set up and 99% of the code will work straight away, however we will also be using some modules that come pre-installed in the colab environment and it means we will save any installation issues for later in the course.

Here we'll just connect our VScode to an online python environment via the extension:
- Press Ctrl+Shift+X (or Cmd+Shift+X)
- Search for "Colab"
- Install the extension by Google

### Connect to Colab Kernel

- Open a the Day1-Introduction notebook (.ipynb file) from the repository
- Click the kernel picker in the top-right corner
- Select "Colab" from the list of available kernels
- Wait for "Connected to Colab" to appear in the status bar

Run the first codeblock to test that it's running!
